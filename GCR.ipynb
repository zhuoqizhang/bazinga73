```python
import treecorr
import GCRCatalogs
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
import math
%matplotlib inline
```

```python
gc = GCRCatalogs.load_catalog('cosmoDC2_v1.1.4_small')
```

```python
data = gc.get_quantities(['ra', 'dec', 'redshift'], filters=['mag_r_lsst<28'])
```

```python
mask_redshift = (data['redshift']>0.2)*(data['redshift']<0.3)
```

```python
plt.hist(data['redshift'][mask_redshift])
plt.xlabel('redshift')
plt.ylabel('count')
plt.title('redshift count distribution')
plt.show()
```

```python
plt.hist(data['redshift'])
plt.xlabel('redshift')
plt.ylabel('count')
plt.title('redshift count distribution')
plt.show()
```

```python
ra_min = min(data['ra'])
ra_max = max(data['ra'])
dec_min = min(data['dec'])
dec_max = max(data['dec'])
print(ra_min, ra_max, dec_min, dec_max)
```

```python
ra_bin = np.linspace(math.ceil(ra_min), math.floor(ra_max), math.ceil(ra_max)-math.floor(ra_min)-1)
dec_bin = np.linspace(math.ceil(dec_min), math.floor(dec_max), math.ceil(dec_max)-math.floor(dec_min)-1)
print(ra_bin)
print(dec_bin)
```

```python
plt.figure()
plt.hist2d(data['ra'], data['dec'], bins=[ra_bin,dec_bin], cmap='PuBu')
plt.colorbar(label='number density (per sq. deg.)')
plt.ylabel('dec')
plt.xlabel('ra')
plt.title('Galaxy distribution')
plt.tight_layout()
```
